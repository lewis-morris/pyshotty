## Import pyshotty

```python
from pyshotty import Firefox
```

### basic example - grab png 

```python
screen = Firefox()
filename = screen.grab("www.google.com")
filename
```

### output as jpg

```python
screen = Firefox()
filename = screen.grab("www.google.com", ret_type="jpg")
filename
```

### output as pil image

```python
screen = Firefox()
filename = screen.grab("www.google.com", ret_type="pil")
filename
```

### change output size

```python
screen = Firefox((540, 1170))
filename = screen.grab("www.google.com", ret_type="pil")
filename
```

### change output folder and file prefix

```python
screen = Firefox(save_path="/home/lewis", file_prefix="my_fab_images")
filename = screen.grab("www.google.com", ret_type="jpg")
filename
```
